```python
# We load and clean the data.
# For total hours and hours per week, we decided to set limits because some of the values were too large to the point of being
# impossible.  We also subset out class '7' because that particular subset did not contain personal information (i.e. Total
# hours played).
fullData= pd.read_csv("SkillCraft1_Dataset.csv")
index = fullData.iloc[:,4] != '?'
myData = fullData.loc[index,:]
index7 = myData.iloc[:,1] != 7
myData = myData.loc[index7,:]
myData = myData.convert_objects(convert_numeric=True)
myData = myData.loc[myData.iloc[:,3] <= 77,:]
myData = myData.loc[myData.iloc[:,4] <= 5000,:]

```

# Trees

```python
# Cross validating for parameters to use to create trees
# finK represents the parameter values to use for each matchup
# finScores represents the accuracy for each matchup
kVals = range(5, 200, 15)
lVals = range(5,200,15)
finK = {}
finScores = {}
for a in itertools.combinations(set(League),2): # All combinations of 2 leagues
    Ind = (League == a[0]) | (League == a[1])
    sData = subData.loc[Ind,:]
    sLeague = League.loc[Ind]
    features = sData.values
    labels = sLeague.values
    LOO = cross_validation.KFold(len(labels),10)
    scores = []
    closeS = []
    interScores = np.zeros((13,13))
    for k in range(0,13): # Decision trees have two parameters, so we do a grid search across both
        for l in range(0,13):
         for trainInd, testInd in LOO:
             Xtrain, Xtest = features[trainInd], features[testInd]
             Ytrain, Ytest = labels[trainInd], labels[testInd]
             clf = tree.DecisionTreeClassifier(min_samples_split=kVals[k],min_samples_leaf=lVals[l],random_state=10)
             clf = clf.fit(Xtrain, Ytrain)
             predVals = clf.predict(Xtest)
             #correctVals = np.sum(np.abs(predVals-Ytest) <= 1)
             #closeS.append(float(correctVals)/len(predVals))
             scores.append(clf.score(Xtest,Ytest))
            
         interScores[l,k] = np.mean(scores)
            
    i,j = np.unravel_index(interScores.argmax(), interScores.shape)    
    finK.update({'%dv%d'%(a[0],a[1]):[i,j]})
    finScores.update({'%dv%d'%(a[0],a[1]):interScores[i,j]})    
        
    
```

```python
# Creating final trees 
# featImp is a list of the important variables
# featImpVals is a list of the values of how important the variables are
featImp = {}
featImpVals = {}
finKK = [v for k,v in finK.items()]
finInd = [k for k,v in finK.items()]
keep = zip(finInd, finKK)
finKS = np.array([e[1] for e in sorted(keep)])
k=0
capInd=0
capBox = ['Fig. 3) This tree features a matchup of the two lowest leagues. ',
          'Note that APM is the first variable the tree splits on.',
          'This tree features a matchup of a low and middle league. We see that APM is ',
          'still present, but ActionLatency has become the first variable the tree splits on.',
          'This tree features a matchup of a middle and high league. Notice how it is more complicated than the previous two trees. ',
          'New variables involve the use of hotkeys.',
          'This tree features a matchup of two high leagues. It is not as complicated as',
          'the previous tree. ActionLatency is still the first variable the tree splits on.'
          ]
for a in itertools.combinations(set(League),2):
 features = subData.loc[(League == a[0]) | (League == a[1]),:]
 labels = League.loc[(League == a[0])|(League == a[1])]
 clf = tree.DecisionTreeClassifier(min_samples_split = kVals[finKS[k][1]], min_samples_leaf = lVals[finKS[k][0]],random_state=5000)
 clf = clf.fit(features.values, labels.values)
 imps = clf.feature_importances_ # important variables
 featImpVals.update({'%dv%d'%(a[0],a[1]):sorted(imps,reverse=True)})
 ColKeep = zip(imps, subData.columns.values)
 SortCols = np.array([e[1] for e in sorted(ColKeep, reverse=True)])
 SortInd = np.array([e[0] for e in sorted(ColKeep, reverse=True)])
 NoZero = SortInd != 0
 T = SortCols[NoZero]
 featImp.update({'%dv%d'%(a[0],a[1]):T})
 # Plot certain matchups   
 if ((a[0]==1)&(a[1]==2))|((a[0]==5)&(a[1]==6))|((a[0]==2)&(a[1]==4))|((a[0]==3)&(a[1]==6)):
  strC = "dot -Tpng SCtest.dot -o " + "%dv%d.png"%(a[0], a[1])
  with open("SCtest.dot", 'w') as f:
    f = tree.export_graphviz(clf, out_file=f,feature_names = subData.columns.values)
  
  os.system(strC)
  os.unlink('SCtest.dot')
  # Manually adding title and captions to the trees
  fontPath = '/usr/share/fonts/truetype/liberation/LiberationMono-Regular.ttf'
  fontP = ImageFont.truetype(fontPath, 40)
  fontP1 = ImageFont.truetype(fontPath, 25)
  newIM = ImageOps.expand(Image.open('%dv%d.png'%(a[0],a[1])),border=100,fill='white')
  draw = ImageDraw.Draw(newIM)
  draw.text((newIM.size[0]/4,50),'%dv%d Matchup Tree'%(a[0],a[1]),font=fontP,fill='black')
  draw.text((10,newIM.size[1]-newIM.size[1]/8),capBox[capInd],font=fontP1,fill='black')
  draw.text((10,newIM.size[1]-newIM.size[1]/12),capBox[capInd+1],font=fontP1,fill='black')
  del draw
  newIM.save('%dv%dWithTitle.png'%(a[0],a[1]))
  close()
  capInd = capInd+2
    
 k = k+1
       
        
        
        
```

# K Nearest Neighbors

```python
predictors = subData
response = League
```

```python
# KNN 
cmap_light = ListedColormap(['#e41a1c', '#377eb8'])
cmap_bold = ListedColormap(['#e41a1c', '#377eb8'])
colBank = ['#e41a1c', '#377eb8']

predictors = subData
response = League

predictors = predictors.loc[response!=3,:]
response = response.loc[response!=3]
predictors = predictors.loc[response!=4,:]
response = response.loc[response!=4]
predictors = predictors.loc[response!=5,:]
response = response.loc[response!=5]
predictors = predictors.loc[response!=6,:]
response = response.loc[response!=6]


a=pylab.rcParams['figure.figsize']
pylab.rcParams['figure.figsize'] = (30.0,20.0)


X=predictors.iloc[:,[10,11]]
y=response
h=0.2
for weights in ['uniform']:
    # we create an instance of Neighbours Classifier and fit the data.
    clf = KNeighborsClassifier(20, weights=weights)
    clf.fit(X, y)

    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, m_max]x[y_min, y_max].
    x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
    y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    fig, ax = plt.subplots(1)
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    # plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, cmap=cmap_bold)
    for i in range(1,3):
        x1New = X.loc[y==i].iloc[:,0]
        x2New = X.loc[y==i].iloc[:,1]
        ax.scatter(x1New,x2New,label=str(i),color=colBank[i-1],marker="8", alpha=1,edgecolor='black',s=100)
                     
    ax.legend(prop={'size':20},loc=4)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("2-Class classification (k = %i, weights = '%s')"
             % (20, weights), fontsize=60)
    plt.xlabel('Gap in PAC',fontsize=60)
    plt.ylabel('Action Latency',fontsize=60)
    plt.figtext(0.1,-0.25,'Fig. 7) The point of this graph is to show you that when you are looking \nat two classes that are right next to'+
                ' each other, then you see a lot of \noverlap in the data. Here, the red color is class 1 (bronze) and the \nblue'+
                ' color denotes class 2 (silver). It would seem that KNN will not \nreally do a good job of accurately being able to'+
                ' distinguish between any \nof the classes that are one apart. This should be pretty obvious, since \nplayers from' +
                ' classes right next to one another probably have similar skill \nlevels, and perhaps have not converged to their true'+
                ' league yet.',fontsize=50) 
    pl.savefig('oneTwo.jpg',bbox_inches='tight')
    pl.show()
    pl.close()

```

```python
cmap_light = ListedColormap(['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33'])
cmap_bold = ListedColormap(['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33'])
colBank = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33']

predictors = subData
response = League

predictors = predictors.loc[response!=2,:]
response = response.loc[response!=2]
predictors = predictors.loc[response!=3,:]
response = response.loc[response!=3]
predictors = predictors.loc[response!=4,:]
response = response.loc[response!=4]
predictors = predictors.loc[response!=5,:]
response = response.loc[response!=5]


a=pylab.rcParams['figure.figsize']
pylab.rcParams['figure.figsize'] = (30.0,20.0)


# KNN for only three classes
X=predictors.iloc[:,[10,11]]
y=response
h=0.2
for weights in ['uniform']:
    # we create an instance of Neighbours Classifier and fit the data.
    clf = KNeighborsClassifier(20, weights=weights)
    clf.fit(X, y)

    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, m_max]x[y_min, y_max].
    x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
    y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    fig, ax = plt.subplots(1)
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    # plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, cmap=cmap_bold)
    for i in [1,6]:
        x1New = X.loc[y==i].iloc[:,0]
        x2New = X.loc[y==i].iloc[:,1]
        ax.scatter(x1New,x2New,label=str(i),color=colBank[i-1],marker="8", alpha=1,edgecolor='black',s=100)
                     
    ax.legend(prop={'size':20},loc=4)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("2-Class classification (k = %i, weights = '%s')"
             % (20, weights),fontsize=60)
    plt.xlabel('Gap in PAC',fontsize=60)
    plt.ylabel('Action Latency',fontsize=60)
    plt.figtext(0.1,-0.25,'Fig. 8) This graph is supposed to be parallel to the one previously, but now we are looking at \ntwo classes furthest'+
                ' apart from each other. You see that this time the boundary \ndoes a much better job of identifying the two. '+ 
                'Here, the red color is class 1 (bronze) \nand the yellow color is class 6 (diamond). Even with the classes being very'+
                ' far \nfrom each other, there are still some data points that are categorized wrong. What this \nsuggests is that perhaps'+
                ' these two variables are not enough to distinguish between \nplayers, and are not a great representative of player skill. '+
                'Our actual model uses \nmore than just these two variables, so it is okay for this to not be perfect at classifying here.',
                fontsize=50) 
    pl.savefig('oneSix.jpg',bbox_inches='tight')
    pl.show()
    pl.close()
    pylab.rcParams['figure.figsize'] = a
```

```python
# Define a KNN function
def KNN(k):
    """k nearest neighbors"""
    scores = []
    closeScores = []
    for train, test in kf:
        trainingSet = predictors.iloc[train,:]
        trainingSetY = response.iloc[train]
        testSet = predictors.iloc[test,:]
        testSetY = response.iloc[test]
        neigh = KNeighborsClassifier(n_neighbors=k)
        neigh.fit(trainingSet, trainingSetY) 
        predictedValues = KNeighborsClassifier.predict(neigh,testSet)
        correct = sum(abs(predictedValues - testSetY) <= 1) 
        closeScores.append(float(correct) / len(testSetY))
        scores.append(KNeighborsClassifier.score(neigh,testSet, testSetY))
    return (np.mean(scores))
predictors=subData
response=League
kf = KFold(len(response), n_folds=10, indices=False)
knnPred=[]
for k in range(1,250):
    knnPred.append(KNN(k))
    
```

```python
neigh = KNeighborsClassifier(n_neighbors=knnPred.index(max(knnPred)))
neigh.fit(predictors, response) 
predictedValues = KNeighborsClassifier.predict(neigh,predictors)
```

```python
z=zip(response,predictedValues)
sz=sorted(collect.Counter(z).items())
sz.append(((1,5), 0))
sz.append(((5,1), 0))
sz.append(((6,1), 0))
sz.append(((6,2), 0))
sz=sorted(sz)
testVals=['%.1f'%(float(x[1])*100/3286) for x in sz]
keys = [x[0] for x in sz]
sz=zip(keys,testVals)
```

```python
a=pylab.rcParams['figure.figsize']
pylab.rcParams['figure.figsize'] = (20.0,10.0)

# Create table of prediction success rate
tabVals=[]
for i in range(1,7):
     tabVals.append([x[1] for x in sz[6*(i-1):(6*i)]])

ax = pl.subplot(111,frame_on=False)
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
theTable = pl.table(cellText=tabVals, rowLabels=['League=%d'%(i) for i in range(1,7)], colLabels=['Pred=%d'%(i) for i in range(1,7)],loc='upper center', colWidths=[0.1]*6)
pl.title('KNN Confusion Matrix',fontsize = 80)
theTable.scale(1.5,3)
thetable.auto_set_font_size(False)
theTable.set_fontsize(25)
pl.figtext(0.05,-0.1,'Table 2. This is a confusion matrix to compare the values of the predicted league compared \nto the current league league of the player. The values inside' +
                'the matrix represent \nthe percentage of the data that is in that category. So this means that any data in the \ndiagonal of the matrix would be where ' + 
                'our method is the same as the current league \nof the player and sums up to about 40% of the data. Also keep in mind that since the \nresponse variable '+
                'is ordered, the level of misclassification is different. For \nexample, if we predict someone plays in the same way that players of league 6 do \n'+
                'but in reality their current league is 1, then we are believing that the current ranking \nis extremely wrong. Since we can assume the current ranking is'+
                'rarely extremely \noff, we would want very few data points to be in the bottom left and top right corners \nof the matrix. This is the case, although we do'+
                'tend to class people as being \nbetter (to the right of the diagonal) than their true league compared to worse (to the left).',fontsize=28) 
pl.show()
pl.savefig('KNNTable.jpg')
pl.close()
pylab.rcParams['figure.figsize']=a
```
