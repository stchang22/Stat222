# PreProcessing

```python
import pandas as pd
from sklearn import tree
from sklearn.externals.six import StringIO
import os
import numpy as np
from sklearn import cross_validation
import itertools
from pylab import *
import Image, ImageDraw, ImageFont, ImageOps
fullData= pd.read_csv("http://summit.sfu.ca/system/files/iritems1/13328/SkillCraft1_Dataset.csv")

index = fullData.iloc[:,4].notnull()
myData = fullData.loc[index,:]
index7 = myData.iloc[:,1] != 7
myData = myData.loc[index7,:]
myData = myData.loc[myData.iloc[:,3] <= 77,:]
myData = myData.loc[myData.iloc[:,4] <= 5000,:]
set(myData.iloc[:,1])
myData
```

```python
League = myData.iloc[:,1]
subData = myData.iloc[:,range(2,20)]
subData
```

# Trees

```python

kVals = range(5, 200, 15)
lVals = range(5,200,15)
finK = {}
finScores = {}
for a in itertools.combinations(set(League),2):
    Ind = (League == a[0]) | (League == a[1])
    sData = subData.loc[Ind,:]
    sLeague = League.loc[Ind]
    features = sData.values
    labels = sLeague.values
    LOO = cross_validation.KFold(len(labels),10)
    scores = []
    closeS = []
    interScores = np.zeros((13,13))
    for k in range(0,13):
        for l in range(0,13):
         for trainInd, testInd in LOO:
             Xtrain, Xtest = features[trainInd], features[testInd]
             Ytrain, Ytest = labels[trainInd], labels[testInd]
             clf = tree.DecisionTreeClassifier(min_samples_split=kVals[k],min_samples_leaf=lVals[l])
             clf = clf.fit(Xtrain, Ytrain)
             predVals = clf.predict(Xtest)
             correctVals = np.sum(np.abs(predVals-Ytest) <= 1)
             closeS.append(float(correctVals)/len(predVals))
             scores.append(clf.score(Xtest,Ytest))
            
         interScores[l,k] = np.mean(scores)
            
    i,j = np.unravel_index(interScores.argmax(), interScores.shape)    
    finK.update({'%dv%d'%(a[0],a[1]):[i,j]})
    finScores.update({'%dv%d'%(a[0],a[1]):interScores[i,j]})    
        
    
```

```python
featImp = {}
featImpVals = {}
finKK = [v for k,v in finK.items()]
finInd = [k for k,v in finK.items()]
keep = zip(finInd, finKK)
finKS = np.array([e[1] for e in sorted(keep)])
k=0
for a in itertools.combinations(set(League),2):
 features = subData.loc[(League == a[0]) | (League == a[1]),:]
 labels = League.loc[(League == a[0])|(League == a[1])]
 clf = tree.DecisionTreeClassifier(min_samples_split = kVals[finKS[k][1]], min_samples_leaf = lVals[finKS[k][0]])
 clf = clf.fit(features.values, labels.values)
 imps = clf.feature_importances_
 featImpVals.update({'%dv%d'%(a[0],a[1]):sorted(imps,reverse=True)})
 ColKeep = zip(imps, subData.columns.values)
 SortCols = np.array([e[1] for e in sorted(ColKeep, reverse=True)])
 SortInd = np.array([e[0] for e in sorted(ColKeep, reverse=True)])
 NoZero = SortInd != 0
 T = SortCols[NoZero]
 featImp.update({'%dv%d'%(a[0],a[1]):T})
 if ((a[0]==1)&(a[1]==2))|((a[0]==5)&(a[1]==6))|((a[0]==2)&(a[1]==4))|((a[0]==3)&(a[1]==6)):
  strC = "dot -Tpng SCtest.dot -o " + "%dv%d.png"%(a[0], a[1])
  with open("SCtest.dot", 'w') as f:
    f = tree.export_graphviz(clf, out_file=f,feature_names = subData.columns.values)
  
  os.system(strC)
  os.unlink('SCtest.dot')
  fontPath = 'LiberationMono-Regular.ttf'
  fontP = ImageFont.truetype(fontPath, 40)
  newIM = ImageOps.expand(Image.open('%dv%d.png'%(a[0],a[1])),border=300,fill='white')
  draw = ImageDraw.Draw(newIM)
  draw.text((200,200),'TITLE',font=fontP,fill='black')
  del draw
  newIM.save('%dv%dWithTitle.png'%(a[0],a[1]))
  close()
    
 k = k+1
        
        
        
        
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
LogitNumVars={}
c = np.linspace(0.01, 1, 20)
for a in itertools.combinations(set(League),2):
     features = subData.loc[(League == a[0]) | (League == a[1]),:].values
     labels = League.loc[(League == a[0])|(League == a[1])].values
     LOO = cross_validation.KFold(len(labels),10)
     logitScores = []
     interScores = []
     for cVals in c:
            for trainInd, testInd in LOO:
                 Xtrain, Xtest = features[trainInd,:], features[testInd,:]
                 Ytrain, Ytest = labels[trainInd], labels[testInd]
                 logit = LogisticRegression(C = cVals, penalty = 'l1')
                 logit = logit.fit(Xtrain, Ytrain)
                 interScores.append(logit.score(Xtest,Ytest))
                    
            logitScores.append(np.mean(interScores))
            
     i = np.argmax(logitScores)
     LogitNumVars.update({'%dv%d'%(a[0],a[1]): [c[i],logitScores[i]]})
            
            
```

```python
LogitNumVars
```

```python
Tic=[k for k,v in LogitNumVars.items()]
Toc=[v[0] for k,v in LogitNumVars.items()]
CVALS = np.array([e[1] for e in sorted(zip(Tic,Toc))])
np.array([e[0] for e in sorted(zip(Tic,Toc))])
CVALS
```

```python
k = 0
LogitVars = {}
np.set_printoptions(precision=3)
for a in itertools.combinations(set(League),2):
     features = subData.loc[(League == a[0]) | (League == a[1]),:]
     labels = League.loc[(League == a[0])|(League == a[1])]
     X_new = LogisticRegression(C = CVALS[k], penalty='l1')
     X_new = X_new.fit(features,labels)
     co = np.round(X_new.coef_[0],3)
     co1 = []
     for i in co:
             co1.append('%.3f'%i)
     colke = zip(np.abs(co), features.columns.values)
     redun = zip(np.abs(co),co1)
     colNa = [e[1] for e in sorted(colke,reverse=True)]
     colD = [e[1] for e in sorted(redun,reverse=True)]
     keep = zip(colNa,colD)
     LogitVars.update({'%dv%d'%(a[0],a[1]):keep})
     k = k+1
```

```python
LogitVars
```

```python
LLab = [k for k,v in LogitVars.items()]
LImpVal = [v for k,v in LogitVars.items()]
keepLImp = zip(LLab, LImpVal)
featLVars = [v[1][:3] for v in sorted(keepLImp)]
featLMatchs = [v[0] for v in sorted(keepLImp)]               
ax1 = subplot(111,frame_on=False)
ax1.xaxis.set_visible(False)
ax1.yaxis.set_visible(False)
theTable = table(cellText = featLVars, rowLabels = featLMatchs, loc='upper center',colWidths=[0.25]*3)
title('Some Logit Coefficients for Each Matchup')
theTable.auto_set_font_size(False)
theTable.set_fontsize(10)
theTable.scale(1.5,1.5)
figtext(0.01,0.02,'Table 2.  This table represents the three largest (in magnitude) coefficients\n given by logistic regression per matchup. '+
        'For the lower leagues,\n we see that APM and ActionsInPAC are prevalent in each matchup.\n'+
        'As we increase the league levels, we see that UniqueHotKeys\n becomes more prevalent along with ActionsinPAC.')
#show(theTable)
savefig('LogitTableWithCap.jpg')
close()
```
